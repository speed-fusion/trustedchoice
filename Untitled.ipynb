```python
final_data = []

for item in all_data:
    trusted_choice_url = "https://www.trustedchoice.com" + item["company_link"]
    agency_website = item["agency_website"]
    data = item["data"]
    agency_name = data["name"]
    
    agency_telephone = data.get("telephone",None)
    
    agency_email = data.get("email",None)
    
    address = data["address"]
    
    street = address["streetAddress"][0]
    locality = address["addressLocality"]
    region = address["addressRegion"]
    postal_code = address.get("postalCode",None)
    
    founding_date = data.get("foundingDate",None)
    
    auto_insurance = item["auto_insurance"]
    home_insurance = item["home_insurance"]
    
    tmp = {
        "agency_name":agency_name,
        "agency_telephone":agency_telephone,
        "agency_email":agency_email,
        "agency_website":agency_website,
        "trusted_choice_url":trusted_choice_url,
        "street":street,
        "locality":locality,
        "region":region,
        "postal_code":postal_code,
        "founding_date":founding_date,
        "auto_insurance":auto_insurance,
        "home_insurance":home_insurance
    }
    
    final_data.append(tmp)
```
